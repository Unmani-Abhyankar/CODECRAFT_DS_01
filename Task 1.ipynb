```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Using the provided dataset
```

```python
## importing the datatset

df = pd.read_csv(r"D:\\API_SP.POP.TOTL_DS2_en_csv_v2_3515.csv",skiprows=4)
```

```python
df.head()
```

```python
df_2024 = df[["Country Name","2024"]].dropna()
```

```python
# Sorting the countries by 2024 population in descending order

df_sorted = df_2024.sort_values(by='2024', ascending=False)
```

```python
top_10_country = df_sorted.head(10)
top_10_country 
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_country, x='2024', y='Country Name')

plt.title('Top 10 Most Populous Countries (2024)', fontsize=14)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Country', fontsize=12)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
##  Histogram â€“ Population Distribution -

plt.hist(df_2024['2024'], bins=20, color='purple', edgecolor='green')

plt.title('Population Distribution Across Countries (2024)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```

```python
# Line Plot:-

# Population trend of India from 1960 to 2022

india = df[df['Country Name'] == 'India'].iloc[0, 4:-1] 
years = india.index.astype(str)
population = india.values.astype(float)

plt.figure(figsize=(10, 5))
plt.plot(years, population, marker='o', color='green')
plt.xticks(rotation=65)
plt.title('Population Growth of India (1960â€“2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Stacked Area Plot for 3 Countries (1960â€“2022) -

# Reloading full data with years
df_years = df[df['Country Name'].isin(['India', 'China', 'United States'])]

# Extracting year columns
df_plot = df_years.set_index('Country Name').loc[:, '1960':'2022'].transpose()
df_plot = df_plot.astype(float)

plt.figure(figsize=(12, 6))
df_plot.plot.area()
plt.title('Population Growth of India, China & USA (1960â€“2022)')
plt.ylabel('Population')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```
